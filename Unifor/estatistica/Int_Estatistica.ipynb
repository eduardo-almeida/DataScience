```python
# Função para classificar, calcular valor acumulado e categorizar
def calcular_categoria(df, coluna_valor):
    # Ordenar o DataFrame pela coluna especificada em ordem decrescente
    df_sorted = df.sort_values(by=coluna_valor, ascending=False)

    # Calcular o valor acumulado
    df_sorted['Valor Acumulado'] = df_sorted[coluna_valor].cumsum()

    # Calcular a porcentagem acumulada
    total_valor = df_sorted[coluna_valor].sum()
    df_sorted['Porcentagem Acumulada'] = df_sorted['Valor Acumulado'] / total_valor * 100

    # Classificar em categorias A, B e C
    def categorizar(porcentagem):
        if porcentagem <= 80:
            return 'A'
        elif porcentagem <= 95:
            return 'B'
        else:
            return 'C'

    df_sorted['Categoria'] = df_sorted['Porcentagem Acumulada'].apply(categorizar)

    # Agrupar e contar as ocorrências por categoria
    resultado = df_sorted.groupby('Categoria').agg(
        Total=(coluna_valor, 'sum'),
        Ocorrencias=('Categoria', 'count')
    ).reset_index()

    # Calcular o total de ocorrências
    total_ocorrencias = resultado['Ocorrencias'].sum()

    # Calcular a porcentagem de cada categoria em relação ao total
    resultado['Total %'] = (resultado['Total'] / total_valor) * 100
    resultado['Ocorrencias %'] = (resultado['Ocorrencias'] / total_ocorrencias) * 100

    # Formatar os valores para duas casas decimais
    resultado['Total'] = resultado['Total'].apply(lambda x: f"{x:.2f}")
    resultado['Total %'] = resultado['Total %'].apply(lambda x: f"{x:.2f}")

    return resultado

# Aplicar a função para as colunas desejadas
colunas_para_analisar = [
    'PIB (Mil Reais)',
    'Impostos (Mil Reais)',
    'VA da agropecuária (Mil Reais)',
    'VA da indústria (Mil Reais)',
    'VA dos serviços (Mil Reais)',
    'VA da administração pública (Mil Reais)'
]

# Gerar os resultados para cada coluna
for coluna in colunas_para_analisar:
    print(f"Resultados para {coluna}:")
    resultado = calcular_categoria(df_final, coluna)
    print(resultado)
    print("\n" + "="*50 + "\n")
```
