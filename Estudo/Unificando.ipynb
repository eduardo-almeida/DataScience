```python
import pandas as pd
```

```python
primeira = pd.read_csv('dados.csv')
segunda = pd.read_csv('segundaFase.csv')
```

```python
primeira.head()
```

```python
segunda.head()
```

```python
df_Total = primeira.merge(segunda,on='matricula', how='right')
```

```python
df_Total.head()
```

```python
df_Total["total"] = pd.to_numeric(df_Total["total"], downcast="integer")
df_Total["nota"] = pd.to_numeric(df_Total["nota"])
#df_Total["nota"] = pd.to_numeric(df_Total["nota"], downcast="float", errors='coerce')
```

```python
df_Total['final'] = df_Total['total']+ df_Total['nota']
```

```python
df_Total.head()
```

```python
df_Total.sort_values('final').to_csv('final.csv')
```
